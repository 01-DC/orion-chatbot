```python
from tkinter import *
root = Tk()
root.title("Chatbot")
def send(event= None):
    send = "You -> "+e.get()
    txt.insert(END, "\n"+send)
    user = e.get().lower()
    for i in range(16):
        if user==df['Questions'][i]:
            txt.insert(END, "\n" + "ORION ->" + df['Answers'][i])
            break
        elif user.find("what is the weather") != -1:
            weather(user.split("in", 1)[1])
            break
    e.delete(0, END)

txt = Text(root)
txt.grid(row=0, column=0, columnspan=2)
e = Entry(root, width=110)
e.grid(row=1, column=0)
root.bind('<Return>', send)
send = Button(root, text="Send", command=send).grid(row=1, column=1)
Button(root, text="Quit", command=root.destroy).grid(row= 1, column= 2)
root.mainloop()
```
